```kotlin
val URL = "jdbc:mariadb://localhost:3306/imdb"
val USER_NAME = "root"
val PASSWORD = "pass"

val dbConfig = DatabaseConfiguration(URL, USER_NAME, PASSWORD)
```

# Convert 10000 rows from _actors_ table to the dataframe

```kotlin
val actorDf = DataFrame.readSqlTable(dbConfig, "actors", 10000)
actorDf
```

# Find top-20 the most popular actor names 

```kotlin
val groupByDF = actorDf.groupBy { first_name } 
groupByDF
```

```kotlin
val count = groupByDF.count()
count
```

```kotlin
val top20ManActorNames = actorDf 
    .groupBy { first_name }
    .count()
    .sortByDesc("count")
    .take(20)
    
top20ManActorNames
```

```kotlin
val top20ManActorNamesPlot = top20ManActorNames.plot {
    bars {
         x(first_name) {
            scale = categorical()
        }
        y(count)
    }
 }
 
 top20ManActorNamesPlot
```
